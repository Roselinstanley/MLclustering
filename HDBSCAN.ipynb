```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset=pd.read_csv("Mall_Customers.csv")
```

```python
x=dataset.iloc[:,[3,4]].values
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
```

```python
from sklearn.cluster import HDBSCAN
HDBS=HDBSCAN(min_cluster_size=8,
                            min_samples=5,
                            cluster_selection_method='eom',
                            allow_single_cluster=False,
                            metric='euclidean',
                            algorithm='kd_tree',
                            leaf_size=5)
clusout=HDBS.fit_predict(x)
```

```python
unsupervised=dataset
unsupervised['cluster group']=clusout
unsupervised
```

```python
import seaborn as sb
facet=sb.lmplot(data=unsupervised,x=unsupervised.columns[3],y=unsupervised.columns[4],hue=unsupervised.columns[5],fit_reg=False,legend=True,legend_out=True)
```

```python
from sklearn.metrics import silhouette_score
ss=silhouette_score(x,clusout)
ss
```
