```python
import pandas as pd
```

```python
dataset=pd.read_csv("Mall_Customers.csv")
```

```python
x=dataset.iloc[:,[3,4]]
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
```

```python
from sklearn.cluster import OPTICS
opt=OPTICS(min_samples=5, xi=0.1, min_cluster_size=0.1)
clusout=opt.fit_predict(x)
```

```python
unsupervised=dataset
unsupervised['clustergroup']=clusout
unsupervised
```

```python
import seaborn as sb
sb.lmplot(unsupervised,x=unsupervised.columns[3],y=unsupervised.columns[4],hue=unsupervised.columns[5],fit_reg=False,legend=True,legend_out=True)
```

```python
from sklearn.metrics import silhouette_score
ss=silhouette_score(x,clusout)
ss
```
